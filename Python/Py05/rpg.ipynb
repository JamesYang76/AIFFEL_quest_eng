```python
# AI
import random

class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        real_damage = damage - self.defense
        if real_damage > 0:
            self.hp -= real_damage

    def roll_damage(self):
        return random.randint(1, self.attack)
```

```python
class Item:
    def use(self, target):
        pass
```

```python
class Potion(Item):
    def __init__(self, heal_amount):
        self.heal_amount = heal_amount

    def use(self, target):
        target.hp += self.heal_amount
        print(f"{target.name}이 포션을 사용해 체력을 {self.heal_amount} 회복했다!")
        print(f"{target.name}의 체력: {target.hp}")

```

```python
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0
        self.inventory = [Potion(30), Potion(30)]

    def gain_experience(self, exp):
        self.experience += exp

    def level_up(self):
        while self.experience >= 500:
            self.experience -= 500
            self.level += 1
            self.attack += 10
            self.defense += 5

    def has_potion(self):
        return len(self.inventory) > 0

    def auto_action(self):
        """
        자동사냥 판단 로직
        """
        if self.hp <= 30 and self.has_potion():
            potion = self.inventory.pop(0)
            potion.use(self)
            return "potion"
        return "attack"
```

```python
class Monster(Character):
    def __init__(self, name, level):
        super().__init__(
            name,
            level,
            random.randint(10, 30) * level,
            random.randint(5, 20) * level,
            random.randint(1, 5) * level
        )
```

```python
class Battle:
    def __init__(self, player, monster):
        self.player = player
        self.monster = monster

    def start(self):
        print(f"\n{self.monster.name} 과의 전투를 시작합니다.")

        while self.player.is_alive() and self.monster.is_alive():
            self.player_turn()
            if self.monster.is_alive():
                self.monster_turn()

        return self.result()

    def player_turn(self):
        action = self.player.auto_action()

        if action == "attack":
            damage = self.player.roll_damage()
            print(f"{self.player.name}이 {self.monster.name}에게 {damage} 만큼 공격했다...!")
            self.monster.take_damage(damage)
            print(f"{self.monster.name}의 체력: {self.monster.hp}")

    def monster_turn(self):
        damage = self.monster.roll_damage()
        print(f"{self.monster.name}이 {self.player.name}에게 {damage} 만큼 공격했다...!")
        self.player.take_damage(damage)
        print(f"{self.player.name}의 체력: {self.player.hp}")

    def result(self):
        if self.player.is_alive():
            self.player.gain_experience(self.monster.level * 20)
            self.player.level_up()
            print("전투 승리!")
            print(f"{self.monster.name} 을 이겼다!")
            return True
        else:
            print("전투 패배...")
            return False
```

```python
def main():
    monster_dict = {
        "슬라임": 1,
        "고블린": 2,
        "오크": 3
    }

    name = input("플레이어 이름을 입력하세요: ")
    player = Player(name)

    for monster_name, level in monster_dict.items():
        monster = Monster(monster_name, level)
        battle = Battle(player, monster)

        if not battle.start():
            print("게임 오버")
            return


if __name__ == "__main__":
    main()
```
