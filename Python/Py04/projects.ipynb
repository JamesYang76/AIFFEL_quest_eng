```python
# 사각형 넓이를 구하는 클래스 완성!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        top, bottom, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시: 윗변,아랫변,높이\n >>>').split(','))
        area = (top + bottom) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result
```

```python
a = Square()
a.rect()
```

```python
a.par()
```

```python
a.trape()
```

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []   # 주문 상품명 저장 리스트
        self.order_price = []  # 주문 가격 저장 리스트
        self.price_sum = 0     # 총 합계 금액
        self.temp = None

    def menu_print(self):
        """메뉴판 출력"""
        for i in range(len(self.menu)):
            print(f"{i+1} {self.menu[i]}  :  {self.price[i]}")

    def _safe_input(self, prompt):
        """숫자 입력 예외 처리를 한 곳에서 수행 (중복 제거)"""
        while True:
            try:
                return int(input(prompt))
            except ValueError:
                print("숫자 번호를 입력해주세요.")

    def _process_order(self, n):
        """온도 선택 및 주문 데이터 업데이트"""
        temp_options = {1: "HOT", 2: "ICE"}
        t = 0
        while t not in temp_options:
            t = self._safe_input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : ")
            self.temp = temp_options.get(t)
            if not self.temp:
                print("1과 2 중 하나를 입력하세요.\n")
        
        name = f"{self.temp} {self.menu[n-1]}"
        cost = self.price[n-1]
        
        self.order_menu.append(name)
        self.order_price.append(cost)
        self.price_sum += cost
        
        return name, cost

    def menu_select(self):
        """주문 선택 (첫 주문 및 추가 주문 루프)"""
        print("\n") 

        # 1. 첫 번째 주문 (유효한 번호 입력까지 반복)
        while True:
            n = self._safe_input("음료를 번호를 입력하세요 : ")
            if 1 <= n <= len(self.menu):
                name, cost = self._process_order(n)
                print(f"{name} : {cost} 원")
                break
            print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 2. 추가 주문 루프
        while True:
            print()
            n = self._safe_input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : ")
            
            if n == 0:
                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price)
                print(f"총 합계 금액은 {self.price_sum}원 입니다.")
                break
            
            if 1 <= n <= len(self.menu):
                name, cost = self._process_order(n)
                print(f"추가 주문 음료 {name} : {cost} 원")
                print(f"합계 : {self.price_sum} 원")
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def table(self):
        """주문 내역 테이블 출력"""
        print('⟝' + '-' * 30 + '⟞')
        for _ in range(5):
            print('|' + ' ' * 31 + '|')
            
        for name, price in zip(self.order_menu, self.order_price):
            print(f"| {name} : {price}원")
            
        print('|' + '-' * 31 + '|')
        print(f"| 합계 금액 : {self.price_sum}원")
        
        for _ in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

    def pay(self):
        """결제 방식 선택"""
        print(f"\n지불하실 금액은 {self.price_sum}원입니다.")
        pay_actions = {
            '1': "직원을 호출하겠습니다.",
            '2': "IC칩 방향에 맞게 카드를 꽂아주세요."
        }
        
        while True:
            method = input("현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : ")
            if method in pay_actions:
                print(pay_actions[method])
                break
            print("다시 결제를 시도해 주세요.")
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
